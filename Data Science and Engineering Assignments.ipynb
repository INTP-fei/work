```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#
df = pd.read_csv('titanic.csv')
# 
df= df.drop_duplicates()
# 
df_cleaned = df.dropna(subset=['Age'])
#
if 'age' in df_cleaned.columns:
    df_cleaned = df_cleaned[df_cleaned['age'] % 1 == 0]
# 
plt.figure(figsize=(6, 5))#(,)
sns.stripplot(x='Sex', y='Age', data=df_cleaned, jitter=True, alpha=0.5)#(xy,)
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Age vs Sex (Scatter Plot)')
plt.grid(True)
#
plt.show()
# 
if 'Age' in df.columns:
    plt.figure(figsize=(6, 5))
    sns.boxplot(y=df['Age'])
    plt.title('Boxplot of Age')
    plt.show()
#  
# 
sex_counts = df_cleaned ['Sex'].value_counts()
# 
plt.figure(figsize=(5, 5))
sex_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#FFFF00','#99ff99'])#(,, 90 ,)
plt.title('Gender Distribution (Pie Chart)')
plt.ylabel('')  #  ylabel
plt.show()
# 
avg_fare_by_class = df_cleaned.groupby('Pclass')['Fare'].mean()
# 
plt.figure(figsize=(6, 5))
avg_fare_by_class.plot(kind='line', marker='o', color='#ADD8E6')#(,
plt.title('Average Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.grid(True)#
plt.show()
```
